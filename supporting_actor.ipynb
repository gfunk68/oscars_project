```python
#Import dependencies
import pandas as pd
```

```python
#Import raw data from webpage
raw_df = pd.read_html("https://en.wikipedia.org/wiki/Academy_Award_for_Best_Supporting_Actor")[2]
```

```python
#Create List of unique oscar years
list_of_years = raw_df.Year.unique()
```

```python
#Loop through the data to select only the Best Supporting Actor winners in each year. 
#Best Supporting Actor is always listed first in each Oscar year amongst the nominees.
years = []
supporting_actor=[]
film=[]

for x in list_of_years:
    years.append(raw_df[raw_df["Year"]==x].iloc[0]["Year"])
    supporting_actor.append(raw_df[raw_df["Year"]==x].iloc[0]["Actor"])
    film.append(raw_df[raw_df["Year"]==x].iloc[0]["Film"])
```

```python
bestsupportingactor_df = pd.DataFrame({"Years": years, "Winner Name":supporting_actor, "Film": film})
```
